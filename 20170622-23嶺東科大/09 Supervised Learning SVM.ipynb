## SVM 分類簡單版

```python
x = np.array([[-3,2], [-6,5], [3,-4], [2,-8]])
y = np.array([1, 1, 2, 2])
```

```python
x[:]
```

```python
x[:, 0]
```

```python
x[:, 1]
```

#### 畫圖

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

#### SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

#### 預測結果

```python
x
```

```python
y
```

```python
clf.predict(x)
```

```python
gd = np.array([[i,j] for i in np.arange(-8, 4, 0.6) for j in np.arange(-10, 6, 0.8) ])
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

## Iris 數據庫

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

#### 資料內容

鳶尾花 (Iris) 的數據, 有三類的鳶尾花我們想用 SVM 做分類。

```python
iris.data
```

#### 答案

```python
iris.target
```

#### 只選兩個 features (為了容易畫圖說明)

```python
x = iris.data[:, :2]
y = iris.target
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
x
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

```python
clf.predict(x)
```

```python
y
```

```python
clf.predict(x) - y
```

```python
gd = np.array([[i,j] for i in np.arange(4,8,0.2) for j in np.arange(1.8,4.5,0.2)])
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

#### 呈現學習成果

學出來的用比較透明的顏色, 真實資料用 100% 不透明。

```python
gd = np.array([[i,j] for i in np.arange(4,8,0.1) for j in np.arange(1.8,4.5,0.1)])
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc, alpha=0.4)
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```
